**Perbedaan fungsi linier dan non linier terhadap parameter model**

Jika kita memiliki fungsi berikut, manakah yang termasuk fungsi linier dan non linier jika a, b, atau c adalah parameter model ?

a. $f(x) = a+bx$

b. $f(x) = a + bx + cx^2$

c. $f(x) = \sqrt{ax}$

d. $f(x) = log(cx)$

e. $f(x,y) = ax^2 + bxy + c^2y$

Jika kita memiliki sembarang nilai $x$ dan $y$ , sedangkan nilai parameter model $a$, $b$, dan $c$ yang sudah ditentukan seperti di bawah ini. Kita dapat melakukan plot grafik terhadap semua fungsi tersebut.

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.random.rand(1,100)
y = np.random.rand(1,100)
a = 5
b = 10
c = 15
```

Berikut akan kita berikan jawaban, fungsi mana yang termasuk fungsi linier dan non linier.

1. Linier
2. Linier
3. Non Linier di $a$
4. Non Linier di $c$
5. Linier di $a$, $b$, non linier di $c$

```python
f1 = a+x*b
plt.plot(x,f1, "bo")
plt.show()
```

```python
f2 = a+x*b+(np.square(x)*c)
plt.plot(x,f2, "ro")
plt.show()
```

```python
f3 = np.sqrt(x*a)
plt.plot(x,f3, "yo")
plt.show()
```

```python
f4 = np.log(x*c)
plt.plot(x,f4, "go")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
X, Y = np.meshgrid(x,y)
f5 = (np.square(X))*a + X*Y*b + Y*(c*c)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X,Y,f5)
plt.show()
```
